```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

open R1 (signal1), store R1 sample rate (samplerate1)

```python
file = 'r1.wav'
signal1 , samplerate1 = librosa.load(file,mono=True,sr=None)
```

convert from samples to seconds

```python
length_s = signal1.shape[0] / samplerate1
print(length_s)
```

```python
time=np.arange(0,signal1.shape[0])/samplerate1
```

plot R1

```python
plt.plot(time,signal1)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.title("R1 Waveform")
```

zoom in R1 plot to 15-25 seconds, -0.5 to 0.5 amplitude

```python
plt.plot(time,signal1)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('R1 Waveform')
plt.xlim(15,25)
plt.ylim(-.5,.5)
```

R1 linear spectrogram

```python
R1spectrogram = librosa.amplitude_to_db(np.abs(librosa.stft(signal1)), ref=np.max)
```

```python
librosa.display.specshow(R1spectrogram, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('R1: linear-frequency power spectrogram')
```

R1 logarithmic spectrogram

```python
librosa.display.specshow(R1spectrogram, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('R1: logarithmic-frequency power spectrogram')
```
